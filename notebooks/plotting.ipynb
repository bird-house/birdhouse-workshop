```python
%matplotlib inline
import matplotlib.pylab as plt
```

```python
from netCDF4 import Dataset
```

```python
f = Dataset('http://www.esrl.noaa.gov/psd/thredds/dodsC/Datasets/ncep.reanalysis/surface/air.sig995.2012.nc')
```

```python
air = f.variables['air']
lat = f.variables['lat'][:]
lon = f.variables['lon'][:]
```

```python
plt.imshow(air[0,:,:])
plt.colorbar();
```

```python
air_c = air[:]  - 273.15
```

```python
plt.imshow(air_c[0,:,:])
plt.colorbar()
```
